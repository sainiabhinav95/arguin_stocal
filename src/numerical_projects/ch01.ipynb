### Distributions as Histograms

```python
# Sample 10000 numbers between 0 and 1
sample = np.random.rand(10000)

# Plot the PDF
n_bins = 50
fig = px.histogram(sample, nbins=n_bins, histnorm='probability density')
fig.update_layout(title='PDF of Uniform Distribution', xaxis_title='Value', yaxis_title='Density')
fig.show()

# Plot the CDF
fig = px.histogram(sample, nbins=n_bins, cumulative=True, histnorm='probability density')
fig.update_layout(title='CDF of Uniform Distribution', xaxis_title='Value', yaxis_title='Cumulative Density')
fig.show()
```

```python
# The plots for X^2 
sample_squared = sample ** 2
# Plot the PDF of X^2
fig = px.histogram(sample_squared, nbins=n_bins, histnorm='probability density')
fig.update_layout(title='PDF of X^2', xaxis_title='Value', yaxis_title='Density')
fig.show()

# Plot the CDF of X^2
fig = px.histogram(sample_squared, nbins=n_bins, cumulative=True, histnorm='probability density')
fig.update_layout(title='CDF of X^2', xaxis_title='Value', yaxis_title='Cumulative Density')
fig.show()
```

### Law of Large Numbers

```python
# Strong law of large numbers
np.random.seed(42)
sample_size = 10000

samples = np.random.exponential(scale=1.0, size=sample_size)
n_values = np.arange(1, sample_size + 1)
cum_means = np.cumsum(samples) / n_values

# Plot the value
fig = px.line(x=n_values, y=cum_means, title="Cumulative Mean of Exponential Samples", labels={'x': 'Number of Samples', 'y': 'Cumulative Mean'})
fig.add_hline(y=1.0, line_dash="dash", line_color="red", annotation_text="Expected Value (1.0)", annotation_position="top right")
fig.show()
```

```python
# Weak Law
np.random.seed(42)
def get_empirical_mean_exp(n_samples):
    samples = np.random.exponential(scale=1.0, size=n_samples)
    emp_mean = np.mean(samples)
    return emp_mean

sample_size = 10000
n1 = 100
n2 = 10000

n1_samples = [get_empirical_mean_exp(n1) for _ in range(sample_size)]
n2_samples = [get_empirical_mean_exp(n2) for _ in range(sample_size)]

# Plot PDF and CDF for n1
fig = px.histogram(n1_samples, nbins=n_bins, histnorm='probability density', title=f'PDF of Empirical Means (n={n1})', labels={'value': 'Empirical Mean'})
fig.show()

fig = px.histogram(n1_samples, nbins=n_bins, cumulative=True, histnorm='probability density', title=f'CDF of Empirical Means (n={n1})', labels={'value': 'Empirical Mean'})
fig.show()

# Plot PDF and CDF for n2
fig = px.histogram(n2_samples, nbins=n_bins, histnorm='probability density', title=f'PDF of Empirical Means (n={n2})', labels={'value': 'Empirical Mean'})
fig.show()

fig = px.histogram(n2_samples, nbins=n_bins, cumulative=True, histnorm='probability density', title=f'CDF of Empirical Means (n={n2})', labels={'value': 'Empirical Mean'})
fig.show()  # Print first 5 samples for n2
```

### Central Limit Theorem

```python
np.random.seed(42)
def y_n(n_samples, scale=1.0):
    samples = np.random.exponential(scale=scale, size=n_samples)
    sum_samples = np.sum(samples)
    # Scale is the mean of the exponential distribution, variance is 

    y_n = (sum_samples - n_samples * scale) / (np.sqrt(n_samples) * scale)

    return y_n

n_samples = 10000
n = 100

y_n_samples = [y_n(n, scale=1.0) for _ in range(n_samples)]

# Plot PDF for y_n
fig = px.histogram(y_n_samples, nbins=n_bins, histnorm='probability density', title=f'PDF of y_n (n={n})', labels={'value': 'y_n'})
fig.show()

gaussian_samples = np.random.normal(loc=0.0, scale=1.0, size=n_samples)
# Plot PDF for Gaussian samples
fig = px.histogram(gaussian_samples, nbins=n_bins, histnorm='probability density', title='PDF of Standard Normal Distribution', labels={'value': 'Value'})
fig.show()


```

### Cauchy Random Variables

```python
np.random.seed(42)

def cauchy_inv(x):
    return np.tan(np.pi * (x - 0.5))

n_samples = 10000
uniform_samples = np.random.rand(n_samples)
cauchy_samples = cauchy_inv(uniform_samples)

# Plot Histogram of Cauchy samples
fig = go.Figure(data=[go.Histogram(x=cauchy_samples, histnorm='probability density', xbins=dict(start=-10, end=10, size=0.1))])
fig.update_layout(title='Histogram of Cauchy Distribution (Custom Inverse Transform)', xaxis_title='Value', yaxis_title='Density')
fig.show()

# Let's get directly from numpy
cauchy_samples_np = np.random.standard_cauchy(size=n_samples)
# Plot Histogram of Cauchy samples from numpy
fig = go.Figure(data=[go.Histogram(x=cauchy_samples_np, histnorm='probability density', xbins=dict(start=-10, end=10, size=0.1))])
fig.update_layout(title='Histogram of Cauchy Distribution (NumPy)', xaxis_title='Value', yaxis_title='Density')
fig.show()
```

```python
standard_gaussian_samples = np.random.normal(loc=0.0, scale=1.0, size=n_samples)
# Plot PDF for Standard Gaussian samples
fig = go.Figure(data=[go.Histogram(x=standard_gaussian_samples, histnorm='probability density', xbins=dict(start=-10, end=10, size=0.1))])
fig.update_layout(title='PDF of Standard Normal Distribution', xaxis_title='Value', yaxis_title='Density')
fig.show()
```
